```python
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

```python
stockanalysis = pd.read_csv('allstocks5yr.csv', index_col='date', parse_dates=['date'])
```

```python
aapl = stockanalysis[stockanalysis.name=='AAPL']

aapl
```

```python
aapl[['open', 'close']].plot(figsize = (12, 4))
plt.show
```

```python
aapl['gain_loss'] = aapl['close'] - aapl['open']

aapl
```

```python
stockanalysis['gain_loss']= stockanalysis['close'] - stockanalysis['open']

stockanalysis
```

```python
stockanalysis.info()
```

```python
stockanalysis['rolling_close_mean'] = stockanalysis['close'].rolling(4).mean()

stockanalysis['rolling_close_mean']
```

```python
stockanalysis
```

```python
stockanalysis['rolling_vol_sum'] = stockanalysis['volume'].rolling(1).sum()

stockanalysis
```
